## Computing Innovations: Benefits & Risks

### Beneficial Effects
- **Medical Advancements**: Better diagnostics, robotic surgeries, and health data usage  
- **Business Efficiency**: Automation and data insights improve productivity  
- **Art & Media**: New creative platforms and global collaboration  
- **Everyday Convenience**: Drones used for deliveries and aerial shots  

### Harmful Effects
- **Cyberbullying**: Leads to anxiety, depression, and isolation  
- **Loss of Privacy**: Data misuse and surveillance issues  
- **Tech Dependency**: Decreases independent thinking  
- **Job Displacement**: Automation replacing human roles  
- **Drone Concerns**: Safety and privacy violations  

---

## Self-Driving Cars

### Benefits
- Reduces human-error-related accidents  
- Improves traffic flow and reduces congestion  
- Makes transportation more accessible  
- Lowers emissions with efficient routes  

### Risks
- Potential job loss in driving-related fields  
- Systems vulnerable to hacking  
- Raises ethical dilemmas (e.g., who’s responsible in a crash?)  
- High upfront cost for adoption and infrastructure changes

### Positive Effects

- Can boost motivation in controlled use  

### Negative Effects

### Solutions
- Set screen time limits  
- Practice mindfulness

## Intellectual Property (IP)

### Types of Protection
- **Copyright**: Protects original artistic or literary works  
- **Patents**: Covers new inventions (typically lasts ~20 years)  
- **Trademarks**: Safeguards logos, brand names, and slogans  
- **Trade Secrets**: Keeps confidential business info protected  

### Preventing Violations
- Use watermarks and DRM tools  
- Register IP officially and defend it legally  
- Educate others on ethical and legal use  

---

## Creative Commons Licenses

- Common types: `CC BY`, `CC BY-SA`, `CC BY-NC`, etc.  
- Allow sharing, remixing, and attribution-based use  
- Useful for educational content, art, and open-source sharing

## Graphs & Heuristics

### Graphs
- Made up of **nodes** (points) and **edges** (connections)  
- Can be directed/undirected, weighted/unweighted  
- Applications: social media networks, maps, recommendation engines  

### Heuristics
- Rules of thumb for solving problems quickly  
- Examples: Greedy algorithm, A* search  
- Common in pathfinding (like Google Maps) and optimization  

---

## Plagiarism in Computing

- **Definition**: Copying code or ideas without giving credit  
- **Consequences**: Legal trouble, academic penalties, damaged reputation  
- **How to Avoid**:
  - Cite all sources  
  - Don’t share code unless allowed  
  - Write your own solutions

## Licensing Overview

- **MIT License**: Very permissive; allows commercial use  
- **GPL**: Requires open-source sharing of modifications  
- **Creative Commons**: For media and educational content  
- **Apache 2.0**: Includes patent protection; flexible for businesses

## Social Network Analysis

- **Nodes** represent users; **edges** represent interactions  
- Used to study influence, friend suggestions, and community clustering  
- Real-world use: Facebook friend graphs, Twitter follower maps

## Decidable vs. Undecidable Problems

- **Decidable**: Always has a definitive answer (e.g., is a number even?)  
- **Undecidable**: No universal solution (e.g., the Halting Problem)  
- **Practical handling**: Software may use timeouts or safeguards to prevent issues  

---

## Random Algorithms

- Use chance to improve fairness, performance, or unpredictability  
- Examples: Cryptographic key generation, simulations, gaming AI  
- In Python: `random.choice()` picks random list elements  

---

## Simulations

- Mimic real-world systems (e.g., disease spread, physics)  
- Used in science, gaming, design testing, etc.  
- Benefits: Cost-effective, low-risk, and scalable  

---

## Base64 Encoding

- Converts binary data into readable text  
- Used in: Email attachments, web APIs, image embedding  
- Tradeoff: Increases file size (~33%) and lacks encryption  

---

## Binary Search

- Efficient method to find items in sorted lists  
- Cuts search space in half each step (O(log n) complexity)  
- Used in databases, autocomplete, and AI decisions  

---

## Logic Gates

- Types: `AND`, `OR`, `NOT`, `NAND`, `NOR`, `XOR`, `XNOR`  
- Form the foundation of computer logic and decision-making  
- Common uses: digital circuits, alarm systems, pattern recognition  

---

## Lists & Filtering

- **Lists**: Ordered, changeable collections of items  
- **Filtering**: Picks specific items that match a condition  
- Time complexity: O(n) for basic filtering  

---

## Binary (Base-2)

- Represents all digital data using only 0 and 1  
- Important for circuits, arithmetic, and data storage  
- Includes operations like binary addition and decimal conversion  

---

## Color Codes

- **Hex**: e.g., `#FF0000` (red)  
- **RGB/RGBA**: e.g., `rgb(255,0,0)`  
- Widely used in web design, apps, video editing, and games
